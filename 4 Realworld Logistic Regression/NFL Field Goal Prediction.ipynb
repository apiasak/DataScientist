* Blocked    (1 if Blocked, 0 ow)

```python
sns.countplot(x='GOOD', hue='homekick', data=players)
```

```python
X = players[['qtr', 'down','togo','ydline','distance','homekick','kickdiff','timerem','offscore','defscore']]
y = players['GOOD']
```

```python
sns.pairplot(players[['qtr', 'down','togo','ydline','distance','homekick','kickdiff','timerem','offscore','defscore','GOOD']],hue='GOOD',palette='bwr')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

** Train and fit a logistic regression model on the training set. **

```python
from sklearn.linear_model import LogisticRegression
```

```python
logm = LogisticRegression()
```

```python
logm.fit(X_train, y_train)
```

```python
# Predict the test datset
predictions = logm.predict(X_test)
```

```python
# Using the classification_report to evaulate the model
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predictions))
```

```python
# Using confusion matrix to evaulate the model
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, predictions)
```

```python
# Check the accuracy of the model
from sklearn import metrics
print(metrics.accuracy_score(y_test, predictions))
```

```python
logm.coef_
```

```python
# Generate table of predictions vs actual
pd.crosstab(y_test, predictions, rownames=['True'], colnames=['Predicted'], margins=True)
#pd.crosstab(y_test, predictions)
```

- 3 True negatives (TN): player correctly identified as non-goal.
- 46 False negatives (FN): kickers the model identified as non-goal who actually did a score.
- 4 False positives (FP): people for whom the model predicted a goal who did not actually score.
- 290 True positives (TP): people the model predicted to score who actually did a score.

```python
conf = confusion_matrix(y_test, predictions)
plt.imshow(conf, cmap='binary', interpolation='None')
plt.show()
```
