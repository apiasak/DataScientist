# Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv("ex1data2.txt",header=None, usecols=[0,2], names=['area','price'])
```

```python
data.head()
```

```python
plt.scatter(data['area'], data['price'])
```

```python
# Calculate r-squared
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(data['area'], data['price'])
```

```python
# Print the r-squared value
r_value ** 2
```

```python
# Plot the data
def predict(x):
    return slope * x + intercept

fitLine = predict(data['area'])

plt.scatter(data['area'], data['price'])
plt.plot(data['area'], fitLine, c='r')
plt.show()
```

```python
# Estimate the price of a 1650 sq-ft
predict(1650)
```

# Linear Regression with sklearn

```python
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
```

## Explore Data

```python
# Random re-order the data 
data = data.sample(frac=1)
```

```python
data.head()
```

```python
# Plot the data
plt.scatter(data['area'], data['price'])
```

## Split Training And Test Datasets

```python
X = data['area']
y = data['price']
```

```python
# Create test dataset from the first 10 records
X_test = X[0:10].reshape(-1,1)
y_test = y[0:10]

# Create our training dataset from the remaining
X_train = X[10:].reshape(-1,1)
y_train = y[10:]
```

### Train The Linear Model

```python
# Create an object that is an ols regression
lim = linear_model.LinearRegression()
```

```python
# Train the model using our training data
model = lim.fit(X_train, y_train)
```

### View The Results

```python
# View the training model's coefficient
model.coef_
```

```python
# View the R-Squared score
model.score(X_test, y_test)
```

```python
# Run the model on X_test and show the first five results
list(model.predict(X_test)[0:5])
```

```python
# View the first five test Y values
list(y_test)[0:5]
```

```python
# Apply the model we created using the training data
# to the test data, and calculate the RSS (residual sum of squares)
((y_test - model.predict(X_test)) **2).sum()
```

```python
# Calculate the MSE (Mean Squared Error)
np.mean((model.predict(X_test) - y_test) **2)
```

```python
# Estimate the price of a 1650 sq-ft
# Scipy = 293237.21718712733
model.predict(1650)
```

# Linear regression with multiple variables

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("ex1data2.txt",header=None, usecols=[0,1,2], names=['area','room','price'])
sns.pairplot(data)
```

```python
y = data['price']
X = data[['area', 'room']]
```

```python
# Use model_selection.train_test_split from sklearn to split the data into training and testing sets. 
# Set test_size=0.2 and random_state=42
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
# The coefficients
print('Coefficients: \n', lm.coef_)
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
# Plot a histogram of the residuals and make sure it looks normally distributed. 
# Use either seaborn distplot, or just plt.hist().
sns.distplot((y_test-predictions),bins=50);
```

```python
coeffecients = pd.DataFrame(lm.coef_,X.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```

* Holding all other features fixed, a 1 unit increase area is associated with an increase of 138 dollars for price.
* Holding all other features fixed, a 1 unit increase room is associated with an decrease of 124 dollars for price.

```python
y_values = lm.predict(X_test[0:5])
print(y_values)
```

```python
# View the first five test Y values
list(y_test)[0:5]
```

```python
# Create new houses DataFrame to predict the price
new_house_info = [{'area': 1650, 'room': 3}, 
                  {'area': 4000, 'room': 6},
                  {'area': 3000, 'room': 2}]
new_house = pd.DataFrame(new_house_info)
new_house.head()
```

```python
y_values = lm.predict(new_house)
print(y_values)
```
